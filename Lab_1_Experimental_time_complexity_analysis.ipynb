The Sum of Vector Elements

Polynomial (Direct Calculation)

```python
def polynom(vec):
    time_start = datetime.datetime.now().timestamp()
    poly = 0
    for i in range(len(vec)):
        poly += vec[i] * (1.5 ** i)
    time_end = datetime.datetime.now().timestamp()
    exec_time = time_end - time_start
    return exec_time

def calculate_exec_time(n):
    vec = np.random.sample(n)
    exec_time = []
    for _ in range(5):
        exec_time = np.append(exec_time, polynom(vec))
    avg_exec_time = np.mean(exec_time)
    return avg_exec_time

def main_p(n):
    avg_exec_time_array = []
    for i in range(1, n + 1):
        avg_exec_time_array = np.append(avg_exec_time_array, calculate_exec_time(i))
    return(avg_exec_time_array)

if __name__ == '__main__':
    n_p = 1500  #      
    n_p_array = range(1, n_p + 1)
    avg_polynom = main_p(n_p)

    fig, ax = plt.subplots()
    ax.plot(n_p_array, avg_polynom, linewidth=0.5, label = 'Empirical results')
    ax.plot(np.linspace(1, n_p, n_p), sklearn.preprocessing.minmax_scale([x * math.log(x) for x in np.linspace(1, n_p, n_p)], feature_range=(0,0.0006)), color='r', label = 'Approximation of theoretical estimations')
    ax.set_title('Polynomial (Direct Calculation)', fontsize=10)
    ax.set_xlabel('N of Vector Elements', fontsize=7)
    ax.set_ylabel('Average Execution Time (Seconds)', fontsize=7)
    ax.tick_params(axis='both', which='major', labelsize=6)
    ax.legend(loc = 2)

    plt.savefig('poly_dir_function.png', dpi = 200, bbox_inches="tight")
    files.download('poly_dir_function.png')
```

Bubble Sort

```python
def bubble_sort(v):
    start = datetime.datetime.now().timestamp()
    for i in range(len(v) - 1):
        for j in range(len(v) - i - 1):
            if v[j] > v[j + 1]:
                v[j], v[j + 1] = v[j + 1], v[j]
    end = datetime.datetime.now().timestamp()
    diff_time = end - start
    return diff_time

def time_searching_bs(n):
    vec = np.random.sample(n)
    diff_time = []
    for _ in range(5):
        diff_time = np.append(diff_time, bubble_sort(vec))
    avg_diff_time = np.mean(diff_time)
    return avg_diff_time

def main_bs(n):
    avg_diff_time_array = []
    for i in range(1, n + 1):
        avg_diff_time_array = np.append(avg_diff_time_array, time_searching_bs(i))
    return avg_diff_time_array

if __name__ == '__main__':
    n = 700
    n_array = range(1, n + 1, 1)
    avg_bs = main_bs(n)

    fig, ax = plt.subplots()
    ax.plot(n_array, avg_bs, linewidth=0.5, label = 'Empirical results')
    ax.plot(np.linspace(1, n, n), sklearn.preprocessing.minmax_scale([x**2 for x in np.linspace(1, n, n)], feature_range=(0,0.07)), color='r', label = 'Approximation of theoretical estimations')
    ax.set_title('Bubble Sort', fontsize=10, loc='center')
    ax.set_xlabel('N of Vector Elements', fontsize=7)
    ax.set_ylabel('Average Execution Time (Seconds)', fontsize=7)
    ax.tick_params(axis='both', which='major', labelsize=6)
    ax.legend(loc = 2)

    plt.savefig('bubble_sort_function.png', dpi = 200, bbox_inches="tight")
    files.download('bubble_sort_function.png')
```

Quick Sort

```python
import datetime
import matplotlib.pyplot as plt
import math
import numpy as np
import sklearn.preprocessing
from google.colab import files

def quick_sort(v):
    less = []
    equal = []
    greater = []
    if len(v) > 1:
        pivot = v[0]
        for x in v:
            if x < pivot:
                less.append(x)
            elif x == pivot:
                equal.append(x)
            elif x > pivot:
                greater.append(x)
        return quick_sort(less) + equal + quick_sort(greater)
    else:
        return v

def time_searching_qs(n):
    vec = np.random.sample(n)
    diff_time = []
    for _ in range(5):
        start = datetime.datetime.now().timestamp()
        tmp = quick_sort(vec)
        end = datetime.datetime.now().timestamp()
        diff_time = np.append(diff_time, end - start)
    avg_diff_time = np.mean(diff_time)
    return avg_diff_time

def main_qs(n):
    avg_diff_time_array = []
    for i in range(1, n + 1):
        avg_diff_time_array = np.append(avg_diff_time_array, time_searching_qs(i))
    return avg_diff_time_array

if __name__ == '__main__':
    n = 2000
    n_array = range(1, n + 1, 1)
    avg_qs  = main_qs(n)
    fig, ax = plt.subplots()
    ax.plot(n_array, avg_qs, linewidth=0.5, label = 'Empirical results')
    ax.plot(np.linspace(1, n, n), sklearn.preprocessing.minmax_scale([x * math.log(x) for x in np.linspace(1, n, n)], feature_range=(0,0.0052)), color='r', label = 'Approximation of theoretical estimations')
    ax.set_title('Quick Sort', fontsize=10, loc='center')
    ax.set_xlabel('N of Vector Elements', fontsize=7)
    ax.set_ylabel('Average Execution Time (Seconds)', fontsize=7)
    ax.tick_params(axis='both', which='major', labelsize=6)
    ax.legend(loc = 2)

    plt.savefig('quick_sort_function.png', dpi = 200, bbox_inches="tight")
    files.download('quick_sort_function.png')
```

Matrix Product

```python
def matrix_multiple(n):
    first_matrix = np.random.random((n, n))
    second_matrix = np.random.random((n, n))
    length = len(first_matrix)
    result_matrix = [[0 for _ in range(length)] for _ in range(length)]
    start = datetime.datetime.now().timestamp()
    for i in range(length):
        for j in range(length):
            for k in range(length):
                result_matrix[i][j] += first_matrix[i][k] * second_matrix[k][j]
    end = datetime.datetime.now().timestamp()
    diff = end - start
    return diff

def time_searching(n):
    diff_time = []
    for _ in range(5):
        diff_time = np.append(diff_time, matrix_multiple(n))
    avg_diff_time = np.mean(diff_time)
    return avg_diff_time

def main(n):
    avg_diff_time_array = []
    for i in range(1, n+1):
        avg_diff_time_array = np.append(avg_diff_time_array, time_searching(i))
    return avg_diff_time_array

if __name__ == '__main__':
    n = 100
    n_array = range(1, n + 1, 1)
    avg_matrix_mult = main(n)

    fig, ax = plt.subplots()
    ax.plot(n_array, avg_matrix_mult, linewidth=0.5)
    ax.plot(np.linspace(1, n, n), sklearn.preprocessing.minmax_scale([x**3 for x in np.linspace(1, n, n)], feature_range=(0,0.7)), color='r', label = 'Approximation of theoretical estimations')
    ax.set_title('Matrix Product', fontsize=10, loc='center')
    ax.set_xlabel('N of Vector Elements', fontsize=7)
    ax.set_ylabel('Average Execution Time (Seconds)', fontsize=7)
    ax.tick_params(axis='both', which='major', labelsize=6)
    ax.legend(loc = 2)

    plt.savefig('matrix_prod_function.png', dpi = 200, bbox_inches="tight")
    files.download('matrix_prod_function.png')
```
